```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../00_data/simulation_data_for_EDA")
```

```python
df_climate_mean = df.groupby('climate_zone').mean()
df_climate_mean = df_climate_mean[['Indoor Temperature [C]', 'Average Unmet Cooling Setpoint Difference [C]',\
                                  'Indoor Relative Humidity [%]',"Equipment Electric Power [kWh]",\
                                   "DHW Heating [kWh]", "Cooling Load [kWh]"]]
df_climate_mean['climate_zone'] = df_climate_mean.index
```

```python
df_climate_mean
```

```python
ax = sns.barplot(x="climate_zone", y="Indoor Temperature [C]", data=df_climate_mean)
ax = ax.set_title('Average Indoor Temperature for Each Climate Zone')
plt.savefig("../graphs/Average Indoor Temperature for Each Climate Zone.png")
```

```python
ax = sns.barplot(x="climate_zone", y="Average Unmet Cooling Setpoint Difference [C]", data=df_climate_mean)
ax = ax.set_title('Average Unmet Cooling Setpoint Difference for Each Climate Zone')
plt.savefig("graphs/Average Unmet Cooling Setpoint Difference for Each Climate Zone.png")
```

```python
ax = sns.barplot(x="climate_zone", y="Indoor Relative Humidity [%]", data=df_climate_mean)
ax = ax.set_title('Average Indoor Relative Humidity for Each Climate Zone')
plt.savefig("graphs/Average Indoor Relative Humidity for Each Climate Zone.png")
```

```python
ax = sns.barplot(x="climate_zone", y="Equipment Electric Power [kWh]", data=df_climate_mean)
ax = ax.set_title('Average Equipment Electric Power for Each Climate Zone')
plt.savefig("graphs/Average Equipment Electric Power for Each Climate Zone.png")
```

```python
ax = sns.barplot(x="climate_zone", y="DHW Heating [kWh]", data=df_climate_mean)
ax = ax.set_title('Average DHW Heating for Each Climate Zone')
plt.savefig("graphs/Average DHW Heating for Each Climate Zone.png")
```

```python
ax = sns.barplot(x="climate_zone", y="Cooling Load [kWh]", data=df_climate_mean)
ax = ax.set_title('Average Cooling Load for Each Climate Zone')
plt.savefig("graphs/Average Cooling Load for Each Climate Zone.png")
```

```python
df_climate_season_mean = df.groupby(['climate_zone','Month']).mean().reset_index()
df_climate_season_mean = df_climate_season_mean[['climate_zone',"Month",'Indoor Temperature [C]', 'Average Unmet Cooling Setpoint Difference [C]',\
                                  'Indoor Relative Humidity [%]',"Equipment Electric Power [kWh]",\
                                   "DHW Heating [kWh]", "Cooling Load [kWh]"]]
#df_climate_season_mean['climate_zone'] = df_climate_season_mean.index
```

```python
df_climate_season_mean
```

```python
ax = sns.lineplot(data=df_climate_season_mean, x="Month", y="Indoor Temperature [C]", hue="climate_zone")
ax = ax.set_title('Average Indoor Temperature for Each Climate Zone by each month')
plt.savefig("graphs/Average Indoor Temperature for Each Climate Zone by each month.png")
```

```python
ax = sns.lineplot(data=df_climate_season_mean, x="Month", y="Average Unmet Cooling Setpoint Difference [C]", hue="climate_zone")
ax = ax.set_title('Average Unmet Cooling Setpoint Difference [C] for Each Climate Zone by each month')
plt.savefig("graphs/Average Unmet Cooling Setpoint Difference [C] for Each Climate Zone by each month.png")
```

```python
ax = sns.lineplot(data=df_climate_season_mean, x="Month", y="Indoor Relative Humidity [%]", hue="climate_zone")
ax = ax.set_title('Average Indoor Relative Humidity [%] for Each Climate Zone by each month')
plt.savefig("graphs/Average Indoor Relative Humidity [%] for Each Climate Zone by each month.png")
```

```python
ax = sns.lineplot(data=df_climate_season_mean, x="Month", y="Equipment Electric Power [kWh]", hue="climate_zone")
ax = ax.set_title('Average Equipment Electric Power [kWh] for Each Climate Zone by each month')
plt.savefig("graphs/Average Equipment Electric Power [kWh] for Each Climate Zone by each month.png")
```
